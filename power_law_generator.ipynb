```python
import scipy.stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def drawPowerLawGraph(csv_file_name):
    dataFrame = pd.read_csv(csv_file_name)
    dataFrame.drop(dataFrame.filter(regex="Unname"),axis=1, inplace=True)
    final_df = dataFrame.sort_values(by=['inDegree'],ascending=True)
    x=final_df['inDegree']
    y=final_df['count']
    plt.plot(x,y)
    plt.xlabel('degree')
    plt.ylabel('count')
    plt.title('Graph of csv'+csv_file_name)
    plt.show()
```

```python
drawPowerLawGraph('amazon.graph.large.csv')
drawPowerLawGraph('amazon.graph.small.csv')
drawPowerLawGraph('youtube.graph.large.csv')
drawPowerLawGraph('youtube.graph.small.csv')
drawPowerLawGraph('dblp.graph.large.csv')
drawPowerLawGraph('dblp.graph.small.csv')
drawPowerLawGraph('gnm1.csv')
drawPowerLawGraph('gnm2.csv')
drawPowerLawGraph('gnp1.csv')
drawPowerLawGraph('gnp2.csv')
```
